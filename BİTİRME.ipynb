```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
excel_train=os.path.join("train","_annotations.csv")
excel_train_csv = pd.read_csv(excel_train, sep=";") #csv ; ile ayrıldığı için bunu belirttim. default olarak virgül ayrımı için işlem oluyor çünkü
excel_train_df=pd.DataFrame(excel_train_csv)
excel_train_df = excel_train_df.set_index('filename') #görüntü isimlerini index olarak excelde tutuyoruz
excel_train_df.head(10)


#excel_test=os.path.join("test","_annotations.csv")
#excel_test_csv= pd.read_csv(excel_test)
#excel_test_df=pd.DataFrame(excel_test_csv)
```

```python
train_image_list=os.listdir("train") #görüntülerin isimleri. yani excelde filenameler
test_image_list=os.listdir("test")

train_image_paths = [os.path.join("train", img) for img in train_image_list] #her bir görüntünün yolu bir dizi içinde tutulur
test_image_paths = [os.path.join("test", img) for img in test_image_list]
```

```python
#VERİ ÖN İŞLEME

def crop_images(excel_train_df): #görüntüleri label kordinatlarına göre kırpmak için fonksiyon
 image_cropped_list=[]
 for i in range(0,1000):
  image_path=os.path.join("train",excel_train_df.index[i])
  image=cv2.imread(image_path)
  image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  xmin_value=excel_train_df['xmin'][i]
  xmax_value=excel_train_df['xmax'][i]
  ymin_value=excel_train_df['ymin'][i]
  ymax_value=excel_train_df['ymax'][i]
  image_cropped=image[int(ymin_value):int(ymax_value),int(xmin_value):int(xmax_value)]
  image_cropped_list.append(image_cropped)

 return image_cropped_list

image_cropped_list=crop_images(excel_train_df)

```
